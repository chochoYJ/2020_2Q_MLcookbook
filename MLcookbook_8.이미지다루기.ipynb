# 8.12 모서리 감지하기 : cv2.cornerHarris
* 두 경계선이 교차하는 지점을 감지
    * 다량의 정보를 가진 영역
* 주요 매개변수 3가지
    * block_size : 각 픽셀에서 모서리 감지에 사용되는 이웃 픽셀의 크기
    * aperture : 소벨커널의 크기 (sobel : 경계선 탐지시 사용. 가장자리 검출)
    * free_parameter : 값이 클수록 부드러운 모서리를 구별
* 결과 이미지 : 가능성있는 모서리

* 흑백 이미지 로드
* 모서리 감지 매개변수 설정
* 모서리 부각 : cv2.dilate(detector_responses, None)
* 이외의 부분은 모두 흰색으로 채워넣고
* 흑백으로 출력


* 다른 유사 방법 : Shi-Tomasi 모서리 감지기 ( goodFreaturesToTrack )
* 뚜렷하게 나타난 모서리를 지정된 개수만큼 찾을 수 있음


```python
image_bgr = cv2.imread('/Users/yeonjin/Desktop/airplane.jpeg')
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
#image_gray = np.float32(image_gray)
image_gray
```

```python
image_gray = np.float32(image_gray)
image_gray
```

```python
#모서리 매개변수 설정
block_size =  2
aperture = 29
free_parameter = 0.04 
#모서리 감지
detector_responses = cv2.cornerHarris(image_gray,
                                     block_size,
                                     aperture,
                                     free_parameter)
detector_responses
```

```python
#모서리 부각
detector_responses = cv2.dilate(detector_responses, None)
detector_responses
```

```python
#가능성이 높은 모서리
plt.imshow(detector_responses, cmap = 'gray'), plt.axis('off')
plt.show()
```

```python
#임계값보다 작은 감지 결과만 남기고 흰색으로 표시 
# Q. 큰 결과를 남기는데 왜..?
threshold = 0.02
image_bgr[detector_responses > threshold * detector_responses.max()] 
```

```python
image_bgr[detector_responses > threshold * detector_responses.max()]  = [255,255,255]
```

```python
image_gray =  cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray , cmap = 'gray'), plt.axis('off')
plt.show()
```

```python
### goodFeaturesToTrack
image_bgr = cv2.imread('/Users/yeonjin/Desktop/airplane.jpeg')
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

#감지할 모서리 개수
corners_to_detect = 10
minimum_quality_score = 0.05
minimum_distance = 25

# 모서리 감지
corners = cv2.goodFeaturesToTrack(image_gray,
                                 corners_to_detect,
                                 minimum_quality_score,
                                  minimum_distance)
corners = np.float32(corners)
# 모서리마다 흰 원 그리기
for corner in corners:
    x,y = corner[0]
    cv2.circle(image_bgr,(x,y),10,(255,255,255),-1)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(image_rgb, cmap = 'gray'), plt.axis('off')
plt.show()
```

```python
# 특성값 담는 리스트
features = []
colors = ('r', 'g', 'b')
for i, channel in enumerate(colors):
    histogram = cv2.calcHist([image_rgb], # 이미지
                            [i], # 채널 인덱스
                            None,# 마스크 없음
                            [256], # 히스토그램 크기
                            [0,256] # 범위
                            )
    features.extend(histogram)
    # features 라는 빈 리스트의 맨 마지막에 값이 추가됨
```

```python
#샘플의 특성값으로 벡터를 만들고, 초기 다섯개 출력
observation = np.array(features).flatten()
observation[0:5]
```
